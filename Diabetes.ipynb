```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.datasets import load_breast_cancer

#load Diabetes
brest=load_breast_cancer()
x=brest.data
y=brest.target
#preprocessing
from sklearn.preprocessing import StandardScaler
standard=StandardScaler()
x=standard.fit_transform(x)

#train_test_split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)

#naive_bayes
from sklearn.naive_bayes import GaussianNB
model=GaussianNB()
model.fit(x_train,y_train)
y_pred_train=model.predict(x_train)
y_pred_test=model.predict(x_test)
#import accuracy,recall,percision
from sklearn.metrics import accuracy_score , recall_score , precision_score , confusion_matrix
acc_train_naive=accuracy_score(y_train,y_pred_train)
acc_test_naive=accuracy_score(y_test,y_pred_test)
r_naive=recall_score(y_test,y_pred_test)
p_naive=precision_score(y_test,y_pred_test)
c_naive=confusion_matrix(y_test,y_pred_test)
#print(f'neive=acc_train:{acc_train},acc_test:{acc_test},recall:{r},percision:{p},confusion;{c}')
print('neive')
print(acc_train_naive,acc_test_naive,
      r_naive,p_naive,c_naive)

#KNN
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=4)
knn.fit(x_train,y_train)
y_pred_train=knn.predict(x_train)
y_pred_test=knn.predict(x_test)
#import accuracy,recall,percision
from sklearn.metrics import recall_score , precision_score , accuracy_score , confusion_matrix
acc_train_knn=accuracy_score(y_train,y_pred_train)
acc_test_knn=accuracy_score(y_test,y_pred_test)
r_knn=recall_score(y_test,y_pred_test)
p_knn=precision_score(y_test,y_pred_test)
c_knn=confusion_matrix(y_test,y_pred_test)
#print(f"kNN=acc_train:{acc_train},acc_test:{acc_test},recall:{r},percision;{p},confusion:{c}")
print('kNN')
print(acc_train_knn,acc_test_knn,
      r_knn,p_knn,c_knn)

#Decision Tree
from sklearn.tree import DecisionTreeClassifier
tree=DecisionTreeClassifier(max_depth=8,min_samples_leaf=4,min_samples_split=2)
tree.fit(x_train,y_train)
y_pred_train_tree=tree.predict(x_train)
y_pred_test_tree=tree.predict(x_test)
#Random Forest
from sklearn.ensemble import RandomForestClassifier
random=RandomForestClassifier(n_estimators=100)
random.fit(x_train,y_train)
y_pred_train_random=random.predict(x_train)
y_pred_test_random=random.predict(x_test)
#accuracy,recall,percision for Tree and random forest
acc_train_tree=accuracy_score(y_train,y_pred_train_tree)
acc_test_tree=accuracy_score(y_test,y_pred_test_tree)
acc_train_random=accuracy_score(y_train,y_pred_train_random)
acc_test_random=accuracy_score(y_test,y_pred_test_random)
r_tree=recall_score(y_test,y_pred_test_tree)
r_random=recall_score(y_test,y_pred_test_random)
p_tree=precision_score(y_test,y_pred_test_tree)
p_random=precision_score(y_test,y_pred_test_random)
c_tree=confusion_matrix(y_test,y_pred_test_tree)
c_random=confusion_matrix(y_test,y_pred_test_random)
print('dicision-tree')
print(acc_train_tree,acc_test_tree,
      r_tree,p_tree,c_tree)
print('random_forest')
print(acc_train_random,acc_test_random,
      r_random,p_random,c_random)
'''print(f"dicision-tree=acc_train:{acc_train_tree},acc_test:{acc_test_tree},recall:{r_tree},percision:{p_tree},confusion:{c_tree},
      random_forest=acc_train:{acc_train_random},acc_test:{acc_test_random},recall:{r_random},percision:{p_random},confusion;{c_random}")'''


#SVM with linear
from sklearn import svm
model=svm.SVC(kernel='linear')
model.fit(x_train,y_train)
y_pred_train=model.predict(x_train)
y_pred_test=model.predict(x_test)
acc_train_linear=accuracy_score(y_train,y_pred_train)
acc_test_linear=accuracy_score(y_test,y_pred_test)
r_linear=recall_score(y_test,y_pred_test)
p_linear=precision_score(y_test,y_pred_test)
c_linear=confusion_matrix(y_test,y_pred_test)

#print(f"svm_linear=acc_train:{acc_train},acc_test:{acc_test},recall:{r},percision:{p},confusion:{c}")
print('svm_linear')
print(acc_train_linear,acc_test_linear,
      r_linear,p_linear,c_linear)

#SVM with rbf
model=svm.SVC(kernel='rbf')
model.fit(x_train,y_train)
y_pred_train=model.predict(x_train)
y_pred_test=model.predict(x_test)
acc_train_rbf=accuracy_score(y_train,y_pred_train)
acc_test_rbf=accuracy_score(y_test,y_pred_test)
r_rbf=recall_score(y_test,y_pred_test)
p_rbf=precision_score(y_test,y_pred_test)
c_rbf=confusion_matrix(y_test,y_pred_test)

#print(f"svm_rbf=acc_train:{acc_train},acc_test:{acc_test},recall:{r},percision:{p},confusion:{c}")
print('svm_rbf')
print(acc_train_rbf,acc_test_rbf,
      r_rbf,p_rbf,c_rbf)

#SVM with sigmoid
model=svm.SVC(kernel='sigmoid')
model.fit(x_train,y_train)
y_pred_train=model.predict(x_train)
y_pred_test=model.predict(x_test)
acc_train_sigmoid=accuracy_score(y_train,y_pred_train)
acc_test_sigmoid=accuracy_score(y_test,y_pred_test)
r_sigmoid=recall_score(y_test,y_pred_test)
p_sigmoid=precision_score(y_test,y_pred_test)
c_sigmoid=confusion_matrix(y_test,y_pred_test)

#print(f"svm_sigmoid=acc_train:{acc_train},acc_test:{acc_test},recall:{r},percision:{p},confusion:{c}")
print('svm_sigmoid')
print(acc_train_sigmoid,acc_test_sigmoid,
      r_sigmoid,p_sigmoid,c_sigmoid)

#logistic
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
model.fit(x_train,y_train)
y_pred_train=model.predict(x_train)
y_pred_test=model.predict(x_test)
acc_train_logistic=accuracy_score(y_train,y_pred_train)
acc_test_logistic=accuracy_score(y_test,y_pred_test)
r_logistic=recall_score(y_test,y_pred_test)
p_logistic=precision_score(y_test,y_pred_test)

#print(f'logistic=acc_train:{acc_train},acc_test:{acc_test},recall:{r},percision:{p}')
print('logistic')
print(acc_train_logistic,acc_test_logistic,
      r_logistic,p_logistic)

#ANN
from sklearn.neural_network import MLPClassifier
model=MLPClassifier(hidden_layer_sizes=64,max_iter=100)
model.fit(x_train,y_train)
y_pred_train=model.predict(x_train)
y_pred_test=model.predict(x_test)
acc_train_ann=accuracy_score(y_train,y_pred_train)
acc_test_ann=accuracy_score(y_test,y_pred_test)
r_ann=recall_score(y_test,y_pred_test)
p_ann=precision_score(y_test,y_pred_test)

print('ANN')
print(acc_train_ann,acc_test_ann,
      r_ann,p_ann)
#print(f'ANN=acc_train:{acc_train},acc_test:{acc_test},recall:{r},percision:{p}')
```
